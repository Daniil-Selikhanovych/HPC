```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load data with results

```python
data = np.loadtxt('ping_pong_v2_res.txt', delimiter = '\t')
data = pd.DataFrame(data, columns=["Size (bytes)", "Iterations", 
                                   "Total time (secs)", 
                                   "Time per message",
                                   "Bandwidth in (Bytes/s)"])
display(data)
```

## Visualize results

```python
fig, ax = plt.subplots()

ax.plot(data["Size (bytes)"], data["Time per message"], 'b', label="time per message")
ax.set_xlabel("Size (bytes)")
ax.set_ylabel("Time per message in seconds")

ax2 = ax.twinx()

ax2.plot(data["Size (bytes)"], data["Bandwidth in (Bytes/s)"], 'r', label="bandwidth")
ax2.set_xlabel("Size in bytes")
ax2.set_ylabel("Bandwidth in (Bytes/s)")

ax.legend()
ax2.legend(loc="lower right")
plt.grid()
plt.show()
```

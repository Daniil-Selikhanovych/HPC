```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filename = "cellular_automata_1d_mpi_res_"
N = str(100)
m=8
plt.figure(figsize=(15,10))
for i in range(1, 7):
    res_0 = []
    res_1 = []
    with open(filename + str(i), 'r') as file:
        lines = file.readlines()
        for line in lines:
            # print(line)
            a, b = line.split("\t")[0], line.split("\t")[1]
            b = b.split("\n")[0]
            res_0.append(int(a))
            res_1.append(float(b))
    res_0 = np.array(res_0)
    res_1 = np.array(res_1)
    plt.plot(res_0, res_1[0]/res_1, '.-', label="N:{}".format(N))
    N = N + str(0)
plt.xlabel("Number of processes")
plt.ylabel("Speedup")
plt.legend()
plt.grid()
plt.show()
```

As we see, speedup > 1 can be achieved only for big values of N >= 10000

## Visualize rule outputs

```python
def visualize(ind_arr, labels, paramater_vals):
    fig, ax = plt.subplots(2, 2, figsize=(10, 10))
    for i in range(len(ind_arr)):
        data_auto = np.loadtxt(f'{ind_arr[i]}res.txt')
        # print(data_auto.shape)
        k = i // 2
        j = i % 2
        ax[k][j].set_title(f"Condition = {labels[i]}, parameter value = {paramater_vals[i]}")
        ax[k][j].imshow(data_auto, cmap='Greys')
        ax[k][j].axis("off")
    plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
names = ["edge", "sharpen", "gaussian", "median"]

plt.figure(figsize=(18, 9))

for i in range(8):
    plt.subplot(2, 4, i + 1)
    if i < 4:
        plt.imshow(Image.open("corgi.jpg"))
        plt.title("Original image")
    else:
        j = i - 4
        plt.imshow(Image.open(names[j] + "_corgi.jpg"))
        plt.title(f"Blurred image, kernel = {names[j]}")
    plt.axis("off")
```
